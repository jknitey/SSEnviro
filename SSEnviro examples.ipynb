```python
# Show weighted sampling, sampling on a map for a field

import folium
from scripts.helper_functions import get_weighted_sample


latitude = 42.133863
longitude = -93.686029

weighted_df = get_weighted_sample(latitude, longitude)

# Create a Folium map centered on the USA
map = folium.Map(location=[latitude, longitude], zoom_start=17)

# Plot station locations
for _, data_point in weighted_df.iterrows():
    lat, lon = data_point['latitude'], data_point['longitude']
    folium.Marker([lat, lon]).add_to(map)
    
## tiles
tile_layer = folium.TileLayer(
    tiles='http://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='ArcGIS World Imagery'
)
tile_layer.add_to(map)
    
# Display the map
map
```
