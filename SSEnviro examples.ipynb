```python
# import functions

from SSEnviro import get_soil_data
from SSEnviro import get_weather_data
from datetime import datetime
```

```python
# pull soil data for the Champaign, IL area

lat = 40.03
long = -88.23

soil_data_df = get_soil_data(id='test1', latitude=lat, longitude=long)
soil_data_df.head()
```

```python
# pull hourly weather data for the Champaign, IL area

lat = 40.03
long = -88.23
start = datetime(2022, 5, 1)
end = datetime(2022, 5, 2)

weather_data_df = get_weather_data(id='test1', type='hourly', latitude=lat, longitude=long, start_date=start, end_date=end)
weather_data_df.head()
```

```python
# pull daily weather data for the Champaign, IL area

lat = 40.03
long = -88.23
start = datetime(2022, 5, 1)
end = datetime(2022, 10, 1)

weather_data_df = get_weather_data(id='test1', type='daily', latitude=lat, longitude=long, start_date=start, end_date=end)
weather_data_df.head()
```
